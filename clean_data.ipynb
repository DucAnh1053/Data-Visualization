```python
import pandas as pd
```

```python
df = pd.read_csv('IT Main in October.csv')
df.head()
```

```python
# Define keywords related to data science
data_science_keywords = ['Data Science', 'Machine Learning', 'AI', 'Artificial Intelligence', 
                         'Deep Learning', 'Data Analyst', 'Data Engineering', 'Big Data']

# Filter jobs where the main skills match the keywords and the country is Vietnam
filtered_jobs = df[
    (df['Country'] == 'Vietnam') &
    (
        df['Main Skill 1'].str.contains('|'.join(data_science_keywords), case=False, na=False) |
        df['Main Skill 2'].str.contains('|'.join(data_science_keywords), case=False, na=False) |
        df['Main Skill 3'].str.contains('|'.join(data_science_keywords), case=False, na=False)
    )
]

# Display the filtered results
print(filtered_jobs.shape)
filtered_jobs.head()
```

```python
# Filter for jobs related to data science in Vietnam
# We'll check the job title and main skills for keywords like 'data', 'data science', 'data analyst', 'data engineer', etc.
data_science_keywords = ['data', 'data science', 'data analyst', 'data engineer', 'machine learning', 'AI', 'artificial intelligence', 'dữ liệu', 'phân tích dữ liệu', 'kỹ sư dữ liệu', 'học máy', 'trí tuệ nhân tạo', 'máy học']

# Filter the jobs where the job title or main skills contain any of the keywords, and the country is Vietnam
df_filtered = df[
    (df['Country'] == 'Vietnam') &
    (
        df['Job Title'].str.contains('|'.join(data_science_keywords), case=False, na=False) |
        df['Main Skill 1'].str.contains('|'.join(data_science_keywords), case=False, na=False) |
        df['Main Skill 2'].str.contains('|'.join(data_science_keywords), case=False, na=False) |
        df['Main Skill 3'].str.contains('|'.join(data_science_keywords), case=False, na=False)
    )
]

# Display the filtered jobs
print(df_filtered.shape)
df_filtered.head()
```

```python
df_filtered.to_csv('IT Main in October_filtered.csv', index=False)
```
